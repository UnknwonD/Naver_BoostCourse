- Îã§Ïùå Ï†úÍ≥µÎêòÎäî ÎîïÏÖîÎÑàÎ¶¨ Í∞íÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Ïà´ÏûêÍ∞Ä ÏùòÎØ∏ÌïòÎäî Î¨∏Ïûê ÌòïÌÉúÎ°ú Î≥ÄÍ≤ΩÌïú Îí§ Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ÏóêÏÑú ÏùåÏ£ºÏôÄ Ìù°Ïó∞Ïóê ÎåÄÌïú ÍµêÏ∞® ÌÖåÏù¥Î∏îÏùÑ Íµ¨Ìï¥ Í∞Å Ìï¥Îãπ ÎπàÎèÑÏàòÎ•º Íµ¨ÌïòÍ≥† ÏãúÍ∞ÅÌôî Ìï¥Ï£ºÏÑ∏Ïöî!

```python
# Ìù°Ïó∞ 1(ÌîºÏö∞ÏßÄ ÏïäÎäîÎã§), 2(Ïù¥Ï†ÑÏóê ÌîºÏõ†ÏúºÎÇò ÎÅäÏóàÎã§), 3(ÌòÑÏû¨ÎèÑ ÌîºÏö∞Í≥† ÏûàÎã§)
# ÏùåÏ£º 0(ÎßàÏãúÏßÄ ÏïäÎäîÎã§), 1(ÎßàÏã†Îã§)
smoke = {1 : "Ìù°Ïó∞ÏïàÌï®", 2: "ÎÅäÏùå", 3: "Ìù°Ïó∞Ï§ë"}
drink = {0: "ÏïàÎßàÏã¨", 1: "ÎßàÏã¨"}
```
