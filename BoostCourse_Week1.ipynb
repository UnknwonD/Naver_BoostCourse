## 기본 자료형 
- List() : ['a', 'b', 'c'] : 일반적인 배열 (벡터)
- Tuple() : ('a', 'b', 'c'), 내부 값을 변경할 수 없는 리스트
- Dict() : {'a' : '에이', 'b' : '비'}, Hash 자료형
- Set() : {'a', 'b', 'c'}, 중복이 존재하지 않음
